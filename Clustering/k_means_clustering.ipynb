# K-Means Clustering

## Importing the libraries

```
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```
dataset = pd.read_csv("Mall_Customers.csv")
X = dataset.iloc[:, [3,4]].values
```

## Using the elbow method to find the optimal number of clusters

```
from sklearn.cluster import KMeans
wcss = []
for i in range(1,11):
  clst = KMeans(n_clusters=i, init='k-means++', random_state = 42)
  clst.fit(X)
  wcss.append(clst.inertia_)
plt.plot(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

## Training the K-Means model on the dataset

```
clst = KMeans(n_clusters=5, init='k-means++', random_state = 42)
y_kmeans = clst.fit_predict(X)
```

```
print(y_kmeans)
```

## Visualising the clusters

```
plt.scatter(X[y_kmeans == 0, 0],X[y_kmeans == 0, 1], s=100, c='red',label='cluster1')
plt.scatter(X[y_kmeans == 1, 0],X[y_kmeans == 1, 1], s=100, c='blue',label='cluster2')
plt.scatter(X[y_kmeans == 2, 0],X[y_kmeans == 2, 1], s=100, c='green',label='cluster3')
plt.scatter(X[y_kmeans == 3, 0],X[y_kmeans == 3, 1], s=100, c='cyan',label='cluster4')
plt.scatter(X[y_kmeans == 4, 0],X[y_kmeans == 4, 1], s=100, c='magenta',label='cluster5')
plt.scatter(clst.cluster_centers_[:,0], clst.cluster_centers_[:,1], s=300, c='yellow', label='centroid')
plt.title("Clusters of customers")
plt.legend()
plt.xlabel("Annual income")
plt.ylabel("Spending score")
plt.show()
```
